```python
# Dot/inner product of 2 Vectors
# No need to explicitly mention as T for Transpose for vectors. It will be automatically calculate as B @ A.T

A = np.array([1,2,3])
B = np.array([4,5,6])
C  = np.dot(A,B)
print("C matrix is (Dot product) = ", C)
```

```python
# Dot/inner product of 2 Vectors
# No need to explicitly mention as T for Transpose for vectors. It will be automatically calculate as B @ A.T

A = np.array([1,2,3])
B = np.array([4,5,6])
C  = A @ B
print("C matrix is (Dot product) = ", C)
```

```python
# Outer product of 2 vectors
# No need to explicitly mention as T for Transpose for vectors. It will be automatically calculate as B @ A.T


A = np.array( [ [1,2,3] ] )
B = np.array([ [4,5,6] ] )
C  = np.outer(A,B)
print("C matrix is (Outer product) = ", C)
```

```python
#Determinant of a matrix
import numpy as np
A = np.array (
    [
        [1,2,2],
        [4,2,4],
        [1,3,2]
    ]
)

det_A = np.linalg.det(A)
print("Determinant of A is = ", det_A)

B = np.array([[1, 2],
              [3, 4]])

print("Determninant of B is = ", np.linalg.det(B))
```

```python
#Inverse of a matrix
A = np.array([[1, 3],
              [3, 5]])
I = np.linalg.inv(A)
print("Inverse of A is = ", I)
```

```python
#Adjoint of a amtrix
# Inverse(A) = 1/det(A) * Adj(A)
# Adj(A) = transpose of co-factos
# Adj(A) = Inverse(A)  *  Det(A)
import numpy as np
A = np.array ( [ [1,2],
                [3,4]
                ])

det_A = np.linalg.det(A)
Inv_A = np.linalg.inv(A)

Adj = Inv_A * det_A
print("Adjoint of A is = ", Adj.T)
```

```python
#Pseudo Inverse of a matrix => A+ = [ Inverse of ( A.T @ A) ] @ A.T
import numpy as np
A= np.array ( [
       [1,2],
    [3,4],
    [5,6]
])
B = (A.T) @ A
print("First part of Formula : " , np.array(B))

inv_B = np.linalg.inv(B)
print("Inverse of B is = ", np.array(inv_B))

Pseudo_B = inv_B @ A.T
print("Pseudo inverse of the matrix is = ", Pseudo_B)

#Now finding it directly:
Pinv_A = np.linalg.pinv(A)
print("Directly finding Pseudo Inverse of A = ",Pinv_A)

```

```python
#Orthonormal of vectos. Its product should be 0 (so, perpendicular) and
#its norm should be 1. so, its orthonormal
import numpy as np
i = np.array( [ [1], [0], [0] ] )
j = np.array ( [ [0], [9], [0] ] )
k = np.array ( [ [0], [0], [1] ] )
orthogonal = 5
orthonormal = 5
if i.T @ j == 0:
  if j.T @ k ==0:
    if k.T @ i ==0:
      print("i,j,k are orthogonal")
      orthogonal = 0

if np.linalg.norm(i, ord=2) == 1:
  if np.linalg.norm(j, ord=2) == 1:
    if np.linalg.norm(k, ord=2) == 1:
      print("i,j,k are orthonormal")
      orthonormal = 1

if orthogonal == 0 and orthonormal == 1:
  print("i,j,k are both orthogonal and orthonormal to eachother")
elif orthonormal !=1:
  print("It is not orthonormal to each other")
elif orthogonal !=0:
  print( 'It is not orthogonal to each other')

```

```python
# svd using the formula

import numpy as np
A = np.array([
    [1,2],
    [3,4],
    [5,6]
])
U, S, VT = np.linalg.svd(A)

print("U:\n", U)

# Create a Sigma matrix with the same dimensions as A
# and place the singular values S on its diagonal.
Sigma = np.zeros((A.shape[0], A.shape[1]))
Sigma[:min(A.shape[0], A.shape[1]), :min(A.shape[0], A.shape[1])] = np.diag(S)
print("Sigma:\n",Sigma)
print("V^T:\n", VT)

# Reconstruct A using U @ Sigma @ VT
print("Reconstructed A =\n", U @ Sigma @ VT)
```
